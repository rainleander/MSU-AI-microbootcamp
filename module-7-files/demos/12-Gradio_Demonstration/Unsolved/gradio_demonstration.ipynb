### Modify the code so we can interact with the application.

```python
# Import gradio 
import gradio as gr
```

```python
# Create a function that takes in a message. 
def run(msg):
    return f"Running with message: {msg}"

app = gr.Interface(fn=run, inputs="text", outputs="text")
app.launch()
```

```python
# Create a function that takes in a message. 
def run(msg):
    return f"Running with message: {msg}"

app = gr.Interface(fn=run, inputs="text", outputs="text")
app.launch(share=True)
```
