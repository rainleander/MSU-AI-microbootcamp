```python
# Define the text to be tokenized.
text = "I love my family. I love my dog"
```

```python
#We can create a dictionary of the words used and assign each word a number.
vocabulary = {}
#Initialize a counter
i = 1 

#Split the words into a list
words = text.split()

#Add each unique word to the vocabulary
for word in words: 
  strip_word = word.strip('.')
  if strip_word in vocabulary:
    continue
  else:
    vocabulary[strip_word]=i
    i+=1

print(vocabulary)
```

```python
# Split the text into sentences. 
sentences = text.split('.')
sentences[0:2]
```

```python
# Create word lists for each sentence.
word_sequences = [sentence.split() for sentence in sentences[0:2]]
word_sequences
```

```python
#Use the vocabulary to encode the words by looking up each word in the vocabulary and replacing the word with its integer. 
sequences =[]
for sequence in word_sequences: 
    int_list = [vocabulary[word] for word in sequence]
    sequences.append(int_list)
    
print(sequences)
```
